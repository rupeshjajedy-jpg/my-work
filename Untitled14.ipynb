```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
```

```python
df = pd.read_excel("/content/WA_Fn-UseC_-Telco-Customer-Churn.xlsm")
```

```python
df
```

```python
df = df.drop(columns=["SrNo"])
```

```python
# Fix TotalCharges
df["TotalCharges"] = pd.to_numeric(df["TotalCharges"], errors="coerce")
df = df.dropna()
```

```python
# Encode target
df["Churn"] = df["Churn"].map({"No": 0, "Yes": 1})
```

```python
train_df, test_df = train_test_split(
    df, test_size=0.2, random_state=42, stratify=df["Churn"]
)
```

```python
def create_subsets(data, n_subsets=5):
    subsets = []
    for _ in range(n_subsets):
        subset = data.sample(frac=1.0, replace=True, random_state=None)
        subsets.append(subset)
    return subsets

subsets = create_subsets(train_df, n_subsets=5)
```

```python
def gini_impurity(y):
    classes, counts = np.unique(y, return_counts=True)
    probs = counts / counts.sum()
    return 1 - np.sum(probs ** 2)
```

```python
def gini_categorical(data, col, target):
    total = len(data)
    gini = 0

    for val in data[col].unique():
        subset = data[data[col] == val]
        weight = len(subset) / total
        gini += weight * gini_impurity(subset[target])

    return gini
```

```python
def gini_numeric(data, col, target):
    values = np.sort(data[col].unique())
    best_gini = float("inf")

    for i in range(1, len(values)):
        threshold = (values[i-1] + values[i]) / 2

        left = data[data[col] <= threshold]
        right = data[data[col] > threshold]

        if len(left) == 0 or len(right) == 0:
            continue

        gini_split = (
            (len(left)/len(data)) * gini_impurity(left[target]) +
            (len(right)/len(data)) * gini_impurity(right[target])
        )

        best_gini = min(best_gini, gini_split)

    return best_gini
```

```python
def best_split(data, target="Churn"):
    gini_scores = {}

    for col in data.columns:
        if col == target:
            continue

        if data[col].dtype == "object":
            gini_scores[col] = gini_categorical(data, col, target)
        else:
            gini_scores[col] = gini_numeric(data, col, target)

    return min(gini_scores, key=gini_scores.get), gini_scores
```

```python
results = []

for i, subset in enumerate(subsets):
    feature, scores = best_split(subset)
    results.append((i+1, feature, scores[feature]))
```

```python
for r in results:
    print(f"Tree {r[0]} → Best Split: {r[1]} | Gini: {round(r[2],4)}")
```
